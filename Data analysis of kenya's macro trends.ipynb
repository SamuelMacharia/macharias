```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
import dash
data = pd.read_csv("file.csv", on_bad_lines="skip", encoding="utf-8", skiprows=3)
data.head()
data_long= data.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name"],
    var_name= "Year",
    value_name="value"
)
#convert year to numeric
data_long["Year"]= pd.to_numeric(data_long["Year"], errors="coerce")

#display cleaned data
data_long.head()

#view all indicators
indicators = data["Indicator Name"].unique()
#print(indicators[:50])

#select relevant indicators
selected_indicators = ["GDP (current US$)",
                       "Inflation, consumer prices (annual %)",
                       "Population, total",
                       "Unemployment, total (% of total labor force) (modeled ILO estimate)",
                       "General government final consumption expenditure (% of GDP)",
                       "Gross capital formation (% of GDP)",
                       "Final consumption expenditure (% of GDP)",
                       "Foreign direct investment, net inflows (% of GDP)"]
filtered_data = data_long[
    (data_long["Indicator Name"].isin(selected_indicators)) &
                          (data_long["Year"]>=2000) & (data_long["Year"]<=2023)]
filtered_data.head(40)
#filtered_data.to_csv("filtered Data.csv", index= False)

#pivot the filtered data for visualization
pivoted_data = filtered_data.pivot(index="Year", columns="Indicator Name", values="value")

#print(pivoted_data.head(10))
#pivoted_data.to_csv("filtered data.csv", index="Year")

#plot each indicator
#pivoted_data.plot(figsize=(10, 6), marker='o')

#plot using seaborn
plt.figure(figsize=(10, 6))
sb.lineplot(data=pivoted_data, palette="tab10", dashes=False, linewidth=2.5)


#add title and labels
plt.title("Trends of Kenya's Economy")
plt.xlabel("Year")
plt.ylabel("value")
plt.yscale('log')
plt.legend(title="Indicator Name")
plt.grid(True)
plt.tight_layout()
plt.show()

#show descriptive statistics
print(pivoted_data.describe())

#skewness and kurtosis
print("Skewness:\n", pivoted_data.skew())
print("Kurtosis:\n", pivoted_data.kurtosis())

#show correlation matrix
corr_matrix = pivoted_data.corr()

#display correlation heatmap using seaborn
sb.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# a multiple linear regression analysis where gdp is the dependent variable
df= pivoted_data

#convert all columns to numbers
df= df.apply(pd.to_numeric, errors="coerce")

#define the independent variables
X = df[["Inflation, consumer prices (annual %)",
                       "Population, total",
                       "Unemployment, total (% of total labor force) (modeled ILO estimate)",
                       "General government final consumption expenditure (% of GDP)",
                       "Gross capital formation (% of GDP)",
                       "Final consumption expenditure (% of GDP)",
                       "Foreign direct investment, net inflows (% of GDP)"]]
#define the dependent variables
y= df["GDP (current US$)"]



#add a constant for the intercept
X=sm.add_constant(X)

#fit the model
mlr_model = sm.OLS(y,X).fit()

#print summary
print(mlr_model.summary())

#print(X.head())
#print(df.dtypes)

#calculate vif
vif_data= pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"]= [vif(X.values, i) for i in range(X.shape[1])]
print(vif_data)


#forecasting
clean_data = pd.read_csv("filtered data.csv")

#check for stationarity
predictors = ["Inflation, consumer prices (annual %)",
                       "Population, total",
                       "Unemployment, total (% of total labor force) (modeled ILO estimate)",
                       "General government final consumption expenditure (% of GDP)",
                       "Gross capital formation (% of GDP)",
                       "Final consumption expenditure (% of GDP)",
                       "Foreign direct investment, net inflows (% of GDP)"]
for i in predictors:
    result = adfuller(clean_data[i].dropna())
    print(f"\n ADF Test For{i}:")
    print(f"ADF Statistic: {result[0]}")
    print(f"P-Value: {result[1]}") # if p<0.05, data is stationary

#forecasting non_stationary data predictors with prophet library
non_stationary_predictors = ["Population, total","General government final consumption expenditure (% of GDP)"]
prophet_forecasts = {}
for i in non_stationary_predictors:
    prophet_data = clean_data[['Year', i]].rename(columns={'Year': 'ds',i:'y'}) #prophet formatn8
    prophet_data['ds']=pd.to_datetime(prophet_data['ds'], format='%Y')
    model= Prophet()
    model.fit(prophet_data)

    #predict the values for the next 10 years
    future= model.make_future_dataframe(periods=10,freq='Y',include_history=False)
    future_years=future['ds'].dt.year.unique()
    forecast = model.predict(future)
    prophet_forecasts[i] = forecast[['ds', 'yhat']].reset_index(drop=True)
    print(f"\n Forecast for {i}:")
    print(forecast.tail())

    model.plot(forecast)

'''

for key in prophet_forecasts:
    prophet_forecasts[key]= prophet_forecasts[key][['ds', 'yhat']].set_index('ds').reindex(future_years)
'''

'''
for key in prophet_forecasts:
    prophet_forecasts[key]=prophet_forecasts[key].loc[future_years]
'''

'''
for key, df in prophet_forecasts.items():
    print(f"{key}: {type(df)}")
'''

#predict the values for stationary predictors using Arima
stationary_predictors = ["Inflation, consumer prices (annual %)",
                        "Unemployment, total (% of total labor force) (modeled ILO estimate)",
                        "Gross capital formation (% of GDP)",
                       "Final consumption expenditure (% of GDP)",
                       "Foreign direct investment, net inflows (% of GDP)"
                       ]
arima_forecast={}
for i in stationary_predictors:
    clean_data[i]=clean_data[i].dropna()
    model = ARIMA(clean_data[i], order=(1,1,1))
    result = model.fit()

    #forecast for 10 years
    arima_forecast[i] = pd.Series(result.forecast(steps=10))
    print(type(arima_forecast))
    print(f"\n forecast for {i}:")
    print(arima_forecast)
    print(arima_forecast.keys())

arima_forecast_df=pd.DataFrame.from_dict({k: v.values for k, v in arima_forecast.items()}, orient="columns")
print(arima_forecast_df.head())
#print(arima_forecast_df.shape())
print(type(arima_forecast_df))


print("Dataframe for arima:", arima_forecast_df.columns)


print(len(future_years))
print(prophet_forecasts.keys())
print(len(prophet_forecasts['Population, total']))
print(len(prophet_forecasts['General government final consumption expenditure (% of GDP)']))
#print(len(arima_forecast.values))

print(type(future_years))
print(type(prophet_forecasts))
print(type(arima_forecast))

#future_years=pd.Series(future_years, name="Year")
print(len(future_years))
prophet_forecasts_df=pd.DataFrame({k: list(v['yhat']) for k, v in prophet_forecasts.items()})

prophet_forecasts_values=prophet_forecasts_df[["Population, total","General government final consumption expenditure (% of GDP)"]]


'''arima_forecasts_values=arima_forecast_df[[
    "Inflation, consumer prices (annual %)",
    "Unemployment, total (% of total labor force) (modeled ILO estimate)",
    "Gross capital formation (% of GDP)",
    "Final consumption expenditure (% of GDP)",
    "Foreign direct investment, net inflows (% of GDP)"
]]

'''
#print("Dataframe for arima columns:", arima_forecasts_values.columns)

#prophet_forecasts_values=prophet_forecasts_values.iloc[:len(future_years)]
print(type(prophet_forecasts_values))

#arima_forecast_values=arima_forecast.to_frame(name="yhat")
#print(arima_forecast_values.columns)

print(prophet_forecasts_df.shape)
print(prophet_forecasts_values.shape)
print(prophet_forecasts_values.columns)
print(len(prophet_forecasts_values['Population, total'].values))
print(len(prophet_forecasts_values['General government final consumption expenditure (% of GDP)'].values))
print(prophet_forecasts_values["General government final consumption expenditure (% of GDP)"].shape)



#predicting gdp with trained mlr
future_predictors=pd.DataFrame({
    "Year": future_years,
    "Inflation, consumer prices (annual %)": arima_forecast_df["Inflation, consumer prices (annual %)"].values,
    "Population, total": prophet_forecasts_values['Population, total'].values,
    "Unemployment, total (% of total labor force) (modeled ILO estimate)": arima_forecast_df["Unemployment, total (% of total labor force) (modeled ILO estimate)"].values,
    "General government final consumption expenditure (% of GDP)": prophet_forecasts_values["General government final consumption expenditure (% of GDP)"].values,
    "Gross capital formation (% of GDP)": arima_forecast_df["Gross capital formation (% of GDP)"].values,
    "Final consumption expenditure (% of GDP)": arima_forecast_df["Final consumption expenditure (% of GDP)"].values,
    "Foreign direct investment, net inflows (% of GDP)": arima_forecast_df["Foreign direct investment, net inflows (% of GDP)"].values



})
future_indicators=future_predictors


print(mlr_model)
print(X.columns)
print(X.dtypes)

#add a constant to match our mlr model
future_predictors= sm.add_constant(future_predictors.drop(columns=['Year']))
print(future_predictors.columns)
print(future_predictors.dtypes)
print(future_predictors.isna().sum())
print(mlr_model.params)
print(X.shape)
print(y.shape)
print(y.dtypes)
#print(np.isinf(future_predictors).sum())




#predict future gdp
future_gdp= mlr_model.predict(future_predictors)
print(future_gdp)

#plotting future gdp for visualization
plt.figure(figsize=(10,5))
sb.lineplot(x=df.index, y=y, label="Actual GDP", marker='o')
sb.lineplot(x=future_years, y=future_gdp, label="Predicted GDP", marker='o', linestyle="dashed")

#labels and titles
plt.xlabel("Year")
plt.ylabel("GDP (current US$)")
plt.title("Actual vs Predicted GDP")
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()


#creating a dashboard with dash
```
